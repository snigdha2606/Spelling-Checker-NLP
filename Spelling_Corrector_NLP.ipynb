```python
#importing all the libraries
import re
import string
from collections import Counter
import numpy as np
```

```python
#reading the corpus
def read_corpus(filename):
  with open(filename, "r") as file:
    lines = file.readlines()
    words = []
    for line in lines:
      words += re.findall(r'\w+', line.lower())

  return words
```

```python
#total number of words in the corpus
words = read_corpus("/Users/snigdhaasthana/Desktop/Assign1_nlp/eng_news_2020_30K-words.txt")
print(f"there are {len(words)} number of words in the corpus")
```

```python
#number of unique words in the corpus
vocab = set(words)
print(f"{len(vocab)} unique words")
```

```python
#frequency of words
word_counts = Counter(words)
print(word_counts["the"])
```

```python
#probability of the word in the corpus
word_count = float(sum(word_counts.values()))
word_probability = {word: word_counts[word] / word_count for word in word_counts.keys()}
```

```python
print(word_probability["the"])
```

```python
#splitting the characters of words for edit operation
def split(word):
  return [(word[:i], word[i:]) for i in range(len(word) + 1)]
```

```python
print(split("the"))
```

```python
#delete edit operation
def delete(word):
  return [l + r[1:] for l,r in split(word) if r]
```

```python
print(delete("the"))
```

```python
#swap edit operation
def swap(word):
  return [l + r[1] + r[0] + r[2:] for l, r in split(word) if len(r)>1]
```

```python
print(swap("the"))
```

```python
string.ascii_lowercase
```

```python
#replace edit operation
def replace(word):
  letters = string.ascii_lowercase
  return [l + c + r[1:] for l, r in split(word) if r for c in letters]
```

```python
print(replace("the"))
```

```python
#insert edit operation
def insert(word):
  letters = string.ascii_lowercase
  return [l + c + r for l, r in split(word) for c in letters]
```

```python
print(insert("the"))
```

```python
#level one edit distance words
#give unique words from all edit operations
def edit_dist1(word):
  return set(delete(word) + swap(word) + replace(word) + insert(word))
```

```python
print(edit_dist1("the"))
```

```python
#level two edit distance words
#give unique words from all edit operations
def edit_dist2(word):
  return set(e2 for e1 in edit_dist1(word) for e2 in edit_dist1(e1))
```

```python
print(edit_dist2("the"))
```

```python
#words in corpus are already correctly spelt
#word and corresponding probability
def correct_spelling(word, vocabulary, word_probabilities):
  if word in vocabulary:
    print(f"{word} is correctly spelt")
    return

  suggestions = edit_dist1(word) or edit_dist2(word) or [word]
  best_guesses = [w for w in suggestions if w in vocabulary]
  return [(w, word_probabilities[w]) for w in best_guesses]
```

```python
#Word correction suggestions
word = "leve"
corrections = correct_spelling(word, vocab, word_probability)

if corrections:
  print(corrections)
  probs = np.array([c[1] for c in corrections])
  best_ix = np.argmax(probs)
  correct = corrections[best_ix][0]
  print(f"{correct} is suggested as correct spelling for {word}")
```

```python
class spelling_Checker(object):

  def __init__(self, corpus_file_path):
    with open(corpus_file_path, "r") as file:           #autocorrect spelling checker
      lines = file.readlines()
      words = []
      for line in lines:
        words += re.findall(r'\w+', line.lower())

    self.vocabs = set(words)
    self.word_counts = Counter(words)
    total_words = float(sum(self.word_counts.values()))
    self.word_probas = {word: self.word_counts[word] / total_words for word in self.vocabs}

  def _level_one_edits(self, word):                               #level one edit distance
    letters = string.ascii_lowercase
    split = [(word[:i], word[i:]) for i in range(len(word) + 1)]           #all edit distance operations
    delete = [l + r[1:] for l,r in split if r]
    swap = [l + r[1] + r[0] + r[2:] for l, r in split if len(r)>1]
    replace = [l + c + r[1:] for l, r in split if r for c in letters]
    insert = [l + c + r for l, r in split for c in letters]

    return set(delete + swap + replace + insert)

  def _level_two_edits(self, word):
    return set(e2 for e1 in self._level_one_edits(word) for e2 in self._level_one_edits(e1))   #level two edit distance

  def check(self, word):
    candidates = self._level_one_edits(word) or self._level_two_edits(word) or [word]
    valid_candidates = [w for w in candidates if w in self.vocabs]
    return sorted([(c, self.word_probas[c]) for c in valid_candidates], key=lambda tup: tup[1], reverse=True) #returns sorted tuple according to probabilty
```

```python
spellcheck = spelling_Checker("/Users/snigdhaasthana/Desktop/Assign1_nlp/eng_news_2020_30K-words.txt")
```
